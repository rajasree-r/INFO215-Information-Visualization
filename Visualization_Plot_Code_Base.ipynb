**Import Visualization Libraries:**

```
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
%matplotlib inline
```

**Read data:**

```
iris = sns.load_dataset("iris")
iris.head()
```

##1. Scatterplot :

By using Pandas library, we can plot charts with '.plot' extension and specify 'kind' of plots. 

```
iris.plot(kind="scatter", x="sepal_length", y="sepal_width")
plt.xlabel('Sepal length', fontsize=12)
plt.ylabel('Sepal width', fontsize=12)
plt.title('Scatterplot',fontsize=16)
plt.show()
```

Another method of using scatterplot is by using 'plt.scatter' and specifying 'x' and 'y' to be plotted.
The primary difference between 'plt.scatter' and 'plt.plot' is that 'plt.scatter' can be used to create scatter plots where the properties of each individual point can be individually mapped to data.

```
plt.scatter(iris.petal_length, iris.petal_width, marker='o',color='green')
plt.xlabel('Petal length', fontsize=12)
plt.ylabel('Petal width', fontsize=12)
plt.title('Scatterplot',fontsize=16)
plt.show()
```

Categorical scatterplot with non-overlapping points is a **swarmplot**.

```
sns.swarmplot(x="species",y="petal_length",data=iris)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Petal length', fontsize=12)
plt.title('Swarmplot',fontsize=16)
plt.show()
```

##2. Pair plot:

Pair plot can be used to visualize the multidimensional relationships among samples.

```
sns.pairplot(iris, hue='species', height=2.5)
plt.show()
```

##3. Box plot

Looking at an individual feature can be done using a boxplot.

```
#Petal length distribution among various species
sns.boxplot(x="species", y="petal_length", data=iris)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Petal length', fontsize=12)
plt.title('Boxplot - petal length for each species',fontsize=15)
```

 Saving the resulting axes as ax each time causes the plot to be shown on top of the previous axes

Further, boxplot can be improved by adding a layer of individual points on top by using Striplot in Seaborn. By using **'jitter=True'**, all points can be viewed. This helps in improving the effectiveness of the visualization.

```
ax = sns.boxplot(x="species", y="petal_length", data=iris)
ax = sns.stripplot(x="species", y="petal_length", data=iris, jitter=True, edgecolor="gray")
plt.xlabel('Species', fontsize=12)
plt.ylabel('Petal length', fontsize=12)
plt.title('Striplot',fontsize=15)
```

```
# removing species column 
new_data = iris[["sepal_length", "sepal_width", "petal_length", "petal_width"]] 
plt.figure(figsize = (10, 7)) 
new_data.boxplot() 
plt.title('Box plot',fontsize=15)
plt.show()
```

##4. Violin plot:

A violin plot combines the benefits of a boxplot and striplot and simplifies them. The denser regions of the data are thicker, and sparser regions of data are thinner in a violin plot.


```
sns.violinplot(x="species", y="petal_length", data=iris, size=6)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Petal length', fontsize=12)
plt.title('Violinplot - petal length for each species',fontsize=15)
plt.show()
```

##5. KDE plot

KDE plots encode the density of observations on one axis with height along the other axis. It provides a more direct interface with easier access to other options when you just want the density estimate.

```
sns.kdeplot(iris.petal_length, shade=True)
plt.show()
```

The bandwidth **(bw)** parameter of the KDE controls how tightly the estimation is fit to the data, much like the bin size in a histogram.

```
x = iris.petal_length
sns.kdeplot(x)
sns.kdeplot(x, bw=.2, label="bw: 0.2") 
sns.kdeplot(x, bw=2, label="bw: 2")
plt.xlabel('petal length', fontsize=12)
plt.show()
```

##6. Bar chart:

A bar chart is used to compare two variables, categories or items. The x-axis represents the categories and are spaced evenly. The y-axis represents the quantity for each category and is drawn as a bar from the baseline to the appropriate level on the y-axis. 

```
plt.bar(iris.species, iris.sepal_length)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Sepal length', fontsize=12)
plt.title('Barchart - Sepal length for each species',fontsize=15)
plt.show()
```

```
 #Loading a new dataset
titanic = sns.load_dataset("titanic")
```

Barcharts show point estimates and confidence intervals.

```
sns.barplot(x="sex", y="survived", hue="class",data=titanic)
plt.xlabel('Sex', fontsize=12)
plt.ylabel('Survived', fontsize=12)
plt.title('Barchart - Survivals in each class',fontsize=15)
plt.show()
```

Plotting a horizontal barchart with 'plot.barh' :

```
titanic['survived'].value_counts().sort_index().plot.barh()
plt.xlabel('Count', fontsize=12)
plt.ylabel('Survived', fontsize=12)
plt.title('Barchart - survived members',fontsize=15)
plt.show()
```

##7. Pointplot :

Pointplot shows point estimates and confidence intervals as rectangular bars.





```
sns.pointplot(x="class", y="survived", hue="sex", data=titanic,palette={"male":"g","female":"m"}, markers=["^","o"],linestyles=["-","--"])
plt.xlabel('Class', fontsize=12)
plt.ylabel('Survived', fontsize=12)
plt.title('Pointplot - Survivals in each class',fontsize=15)
plt.show()
```

```
sns.catplot(x="sex", y="survived", hue="class", kind="point", data=titanic)
plt.xlabel('Sex', fontsize=12)
plt.ylabel('Survived', fontsize=12)
plt.title('Pointplot - Survival rates',fontsize=15)
plt.show()
```

##8. Countplot:

Countplot shows count of observations.

```
sns.countplot(x="deck", data=titanic, palette="Greens_d")
sns.set_style("whitegrid") #Set the matplotlib parameters
plt.title('Countplot',fontsize=15)
plt.show()
```

```
sns.catplot(y="deck", hue="class", kind="count",palette="pastel", edgecolor=".6",data=titanic)
plt.title('Countplot',fontsize=15)
plt.show()
```

##9.Histogram :


A histogram is the most commonly used graph to show frequency distributions. They are different from barcharts in that histograms are used to show distributions of variables while bar charts are used to compare variables. Histograms plot quantitative data with ranges of the data grouped into bins or intervals while bar charts plot categorical data.

```
plt.figure(figsize = (8,6)) 
x = iris["sepal_length"] 
plt.hist(x, bins = 20, color = "green") 
plt.title("Sepal Length",fontsize=15) 
plt.xlabel("Sepal Length in cm") 
plt.ylabel("Count") 
plt.show()
```

```
# Control the number of bins
plt.figure(figsize = (8,6))
sns.distplot(iris["sepal_length"], bins=20, color = "red" )
plt.title("Sepal Length",fontsize=15) 
plt.xlabel("Sepal Length in cm") 
plt.ylabel("Count") 
plt.show()
```

##10. Heatmap:

A Heatmap is a graphical representation of data where the individual values contained in a matrix are represented as colors. Heatmaps are perfect for exploring the correlation of features in a dataset.

```
plt.figure(figsize = (8,6)) 
sns.heatmap(iris.corr(), annot=True)
plt.title("Heatmap", fontsize =16)
plt.show()
```

Seaborn heatmap for correlation between target variable 'Survived' and feature variables in Titanic dataset :

```
plt.figure(figsize = (8,6))
sns.heatmap(titanic.corr(), annot=True, cbar=True, cmap="RdYlGn")
plt.title("Heatmap", fontsize =16)
plt.show()
```

##11.Lineplot :

A line chart or line graph is a type of chart which displays information as a series of data points called ‘markers’ connected by straight line segments. A line chart is often used to visualize a trend in data over intervals of time(  a time series). The line is often drawn chronologically.

```
plt.figure(figsize = (8,6)) 
sns.lineplot(data=iris.drop(['species'], axis=1))
plt.show()
```

##12.Hexbin plot:

Makes a hexagonal binning plot.

```
titanic.plot.hexbin(x='age',y='survived',gridsize=15, cmap = 'Greens')
plt.show()
```

##13. Pie chart:

The usage of a pie chart is to show the relationship of parts out of a whole. It is really difficult for a human to make the difference between the size of similar slices, hence it must be avoided if possible and replaced with barplot. 

```
plt.figure(figsize=(5,5))
titanic.pclass.value_counts().plot(kind='pie')
plt.title("Pie chart of classes in Titanic", fontsize =15)
plt.legend(loc='best')
plt.show()
```

```
#Analyse People who survived and did not survive based on their class

not_survived_class = titanic['pclass'][titanic['survived']==0]
survived_class = titanic['pclass'][titanic['survived']==1]

plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
not_survived_class.value_counts().plot(kind='pie', title='People who didn\'t survive')
plt.legend(loc='upper left')

plt.subplot(1,2,2)
survived_class.value_counts().plot(kind='pie', title='People who survived')
plt.legend(loc='best')
plt.show()
```
